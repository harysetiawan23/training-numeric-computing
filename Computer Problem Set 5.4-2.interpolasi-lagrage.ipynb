# Import Library

```python
from scipy.interpolate import lagrange
from matplotlib import pyplot as plt
import numpy as np
import math
```

## Define Default Polynomial

Default polynomial is polynomial that comes from the task

```python
n = 11
x = np.linspace(0,1,num=n)
y = [0 for i in range(n)];

for index,value in enumerate(x):
    y[index] = math.sqrt(value)
```

## Define Lagrage Function


```python
f = lagrange(x,y)
```

## Calculate Midpoint

```python
midpoints = x[int(n/2 if n%2 ==0 else (n+1)/2)];
x_midpoints = np.array([midpoints])

y_midpoint = f(x_midpoints)
print("For x = {} --> y {}".format(x_midpoints,y_midpoint))
```

## Custom Input

```python
x_custom = int(0)

while not x_custom > 0:
    x_custom = float(input("Please enter x_value (0 - 1) : "))

## Calculate Custom Input
y_custom = f(x_custom)
```

## Display in Graph

```python
fig = plt.figure(figsize = (10,8))
plt.plot(x, y)
plt.scatter(x, y)
plt.scatter(x_midpoints,y_midpoint)
plt.annotate("Midpoint [{}.{}]".format(x_midpoints,y_midpoint),(x_midpoints,y_midpoint,))
plt.scatter(x_custom,y_custom)
plt.annotate("Custom Point [{}.{}]".format(x_custom,y_custom),(x_custom,y_custom))
# plt.plot(x_new, f(x_new), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
